## Step 2: The data

Our observations were generated by running our simulation script (i.e. running 'python generate_table_of_use_cases.py ./testdata/small-datasets/fixed_number_of_mutations/'), and is imported into a matrix

```python
# rows indexed by n, columns by k
Data = np.array(
    [[2,14,39,38,139,300,570,363,1338],
     [3,27,54,111,192,392,567,779,1097],
     [4,40,85,257,501,700,854,2271,2436],
     [5,53,170,291,428,754,1636,1876,1775],
     [6,109,283,520,1079,1858,1807,3015,2258],
     [7,143,293,559,942,2272,3740,2567,10153],
     [8,177,519,728,1108,2656,1551,7081,2323],
     [9,105,529,2088,2830,2755,9039,3235,27632],
     [10,245,1169,1605,710,5874,12384,7507,19799]])

# #The following dataset is for testing-purposes
# Data = np.random.randint(100,size=(20,20))
# for i in range(Data.shape[0]):
#     for j in range(Data.shape[1]):
#         Data[i,j] += 10*float(i*j)**(0.7)

n_list = (2,3,4,5,6,7,8,9,10)
k_list = (0,1,2,3,4,5,6,7,8)

row_labels = ['n = %i'%n for n in n_list]
column_labels = ['k = %i'%k for k in k_list]
```

## Step 3: Plotting

We first set up the poltting environment:

```python
fig, ax = pl.subplots()

# put the major ticks at the middle of each cell
pl.title('$b_{extra} \, = \, 0$')
ax.set_xticks(np.arange(Data.shape[0])+0.5, minor=False)
pl.xticks(rotation=90)
ax.set_yticks(np.arange(Data.shape[1])+0.5, minor=False)
ax.set_xticklabels(row_labels, minor=False)
ax.set_yticklabels(column_labels, minor=False)

heatmap = ax.pcolor(np.transpose(Data), cmap=pl.cm.Greys)
fig.colorbar(heatmap)
```

We export the figure:

```python
savepath = './'
figure_name = 'heatmap_b_extra_0__n_from_2_to_10__k_from_0_to_8'
extensions = ['.pdf','.png','.jpeg','.eps']

for filepath in [savepath+figure_name+ext for ext in extensions]:
    fig.savefig(filepath)
```
