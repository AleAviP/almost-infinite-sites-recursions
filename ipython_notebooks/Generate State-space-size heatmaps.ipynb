## Step 3: Plotting

We first set up the poltting environment:

```python
fig, ax = pl.subplots()

# put the major ticks at the middle of each cell
pl.title('$b_{extra} \, = \, 0$')
ax.set_xticks(np.arange(data_terms_positive_b_0.shape[0])+0.5, minor=False)
pl.xticks(rotation=90)
ax.set_yticks(np.arange(data_terms_positive_b_0.shape[1])+0.5, minor=False)
ax.set_xticklabels(row_labels, minor=False)
ax.set_yticklabels(column_labels, minor=False)

heatmap = ax.pcolor(np.transpose(data_terms_positive_b_0), cmap=pl.cm.Greys)
fig.colorbar(heatmap)
```
